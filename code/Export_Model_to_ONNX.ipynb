# Export model to onnx

```python
import torch
from torch.autograd import Variable
from model_62 import Net
```

```python
# Create the object model
model = Net()

# Load model's parameters
model_checkpoint = '62_classes_acc_8778.pt'
model.load_state_dict(torch.load('models/{0}'.format(model_checkpoint)))

# Export the model
dummy_input = Variable(torch.randn(1, 1, 28, 28),requires_grad=True)
torch.onnx.export(model, dummy_input, "mnist2.onnx")
```
